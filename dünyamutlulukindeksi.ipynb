```python
# Ayn ilemi datanufus iin yap
country_counts_pop = datanufus["Country Name"].value_counts().reset_index()
country_counts_pop.columns = ['Country Name', 'count']
tumulkelerpop = country_counts_pop['Country Name'].tolist()

# Bo bir liste olutur
del_country = []

# tumulkeler listesindeki ve tumulkelerpop listesinde olmayan lkeleri bul
for x in tumulkeler:
    if x not in tumulkelerpop:
        del_country.append(x)

# Sonucu yazdr
print(del_country)
```

```python
yeni_data=data[['Log GDP per capita','year',
       'Social support', 'Healthy life expectancy at birth',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption','region','Life Ladder']].copy()
```

```python
yeni_data.head()
```

```python
# Yeni verilerdeki stun adlarn kontrol etme
print(yeni_data.columns)
print(datanufus.columns)

# Eer stun adlar doru deilse, doru stun adlarn kullanarak dzeltin.
# rnein, 'Country name' yerine 'Country Name' veya 'Year' yerine 'year' gibi.

# 'Country Name' stununu kontrol etme ve filtreleme
if 'Country Name' in yeni_data.columns and 'year' in yeni_data.columns:
    yeni_data = yeni_data[~yeni_data["Country Name"].isin(del_country)]
    yeni_data = yeni_data[~yeni_data.year.isin([2005, 2006, 2007, 2018, 2019, 2020])]
else:
    print("Stun adlar doru deil!")

# Nfus verilerindeki stun adlarn kontrol etme ve pop_dics oluturma
if 'Country Name' in datanufus.columns and 'Year' in datanufus.columns and 'Count' in datanufus.columns:
    pop_dics = {x: {} for x in range(2008, 2018)}
    for i in range(len(datanufus)):
        if datanufus["Year"][i] in range(2008, 2018):
            pop_dics[datanufus["Year"][i]][datanufus["Country Name"][i]] = datanufus["Count"][i]
else:
    print("Nfus verilerindeki stun adlar doru deil!")

```
